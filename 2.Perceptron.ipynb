```python
from sklearn.datasets import make_classification
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#make_classification is used to make datasets
x,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],s=100,c=y,cmap='coolwarm')
#s is the size of the points
#c is the color of the points
#cmap is the color of the points
```

```python
x.shape
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
def perceptron(x,y):
    x=np.insert(x,0,1,axis=1)
    '''x: The array into which values are inserted.
        0: The index before which values are inserted.
        1: The values to insert.
        axis=1: Specifies that the insertion is along columns.
'''
    weights=np.ones(x.shape[1])
    #learning rate
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,100)
        y_hat=step(np.dot(x[j],weights))
        weights=weights+lr*(y[j]-y_hat)*x[j]
    return weights[0],weights[1:]
```

```python
intercept_,coef=perceptron(x,y)
```

```python
print(intercept_,coef)
```

```python
#slope
m=-coef[0]/coef[1]
b=-(intercept_/coef[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=1)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
